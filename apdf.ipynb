# Scrape [Indo-Pacific Defense Forum ](http://apdf-magazine.com)

```python
#export
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm_notebook
import time

data_path= 'data/'
```

## Get Categories

```python
#export
page = requests.get('http://apdf-magazine.com/category/top-stories/')
soup = BeautifulSoup(page.content, 'html.parser')
cats = soup.find('div', id='categories-2').find_all('a')
cat_df = pd.DataFrame({'cat_name':[i.text for i in cats],
                      'cat_url':[i.attrs['href'] for i in cats]})
cat_df
```

```python
#export
max_pages = []
for cat_url in tqdm_notebook(cat_df.cat_url):
    page = requests.get(cat_url)
    time.sleep(5) #time out
    soup = BeautifulSoup(page.content, 'html.parser')
    page_nb = soup.find_all('a', class_='page-numbers')
    if len(page_nb)==0: 
        max_page = 1
    else:
        max_page = int(page_nb[-1].text)
    max_pages.append(max_page)
cat_df['max_page'] = max_pages
```

```python
cat_df.to_csv(f'{data_path}apdf_df.csv',index=False)
cat_df
```
